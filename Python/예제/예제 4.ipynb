        취미 : 독서, 피아노, 드론

```python
class Employee :
    total = 0

    def __init__(self, num=None, name=None, Salary=None, Hobby=None) :
        self.num = num
        self.name = name
        self.Salary = Salary
        self.Hobby = Hobby
        
        Employee.total += 1
        
    def display(self) :
        print('결과 :')
        print('\t사번 :', self.num) 
        print('\t이름 :', self.name)
        print('\t급여 :', self.Salary)
        print('\t취미 :', ', '.join(self.Hobby))
        
        print('---------------------------------------------')
        
    @staticmethod
    def getCount() :
        return Employee.total
    
    def setSalary(self, Salary) :
        self.Salary = Salary
        
    def appendHobby(self, Hobby) :
        self.Hobby.extend(Hobby)
        
    
                
e1 = Employee(1111, "홍길동", 1000000, ["여행", "낚시", "운동"])
e1.display()


e2 = Employee(1112, "임꺽정", 1200000, ["자전거", "볼링"])
e3 = Employee(1113, "신돌석", 1000000, ["독서"])

total = Employee.getCount()
print("총 인원수 : ", total)
    
e3.setSalary(2500000)
e3.appendHobby(["피아노", "드론"])
e3.display()
```
